```python
#источники: 1. "Introduction to Machine Learning with Python: A Guide for Data Scientists"
#               by Andreas C. Muller & Sarah Guido
#           2. "Open Data Science ML course" by Yury Kashnitsky

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
#сгенерирую датасет
X, y = mglearn.datasets.make_forge()
```

```python
#использую логистическую регрессию и линейный метод опорных векторов для классификации
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter( X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("Some feature 0")
    ax.set_ylabel("Some feature 1")
axes[0].legend()
```

```python
#увеличу выборку
rng = np.random.RandomState(3)
X = rng.randn(200, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
```

```python
#применю те же методы
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter( X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("Some feature 0")
    ax.set_ylabel("Some feature 1")
axes[0].legend()
```
