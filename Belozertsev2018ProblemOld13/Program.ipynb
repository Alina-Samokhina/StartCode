## Площадь единичного круга
Оценить площадь единичного круга методом Монте-Карло. Построить график зависимости результата от размера выборки.
Случайным образом выбирается некоторое число точек из единичного квадрата. Берем выборку объемом $10i$, где $i$ принимает значения от 1 до 999, будем строить зависимость величины 4*(количество точек внутри круга)/(объем выборки) от $i$.
Результат должен колебаться около $\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample = []

for i in range (1, 1000):
    
    x = np.random.uniform(0,1,10*i)
    y = np.random.uniform(0,1,10*i)
    s = 4*(sum((x-0.5)**2 + (y-0.5)**2 < 0.25)/(10*i))
    sample.append(s)
```

```python
plt.figure(figsize=(10,8))
plt.plot(sample, 'g')
plt.show()
plt.
```
