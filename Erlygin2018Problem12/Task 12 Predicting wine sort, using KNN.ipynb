## Обработка данных. Выделим вектор ответов и разделим выборку(на 70% обучаем модель на 30 проверяем).

```python
y = data['Type']
```

```python
X = data.drop('Type',axis = 1)
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X, y,test_size = 0.3,random_state = 19)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train);
```

```python
knn_pred = knn.predict(X_valid)
accuracy_score(y_valid, knn_pred)
```

## Проведем настройку гиперпараметров нашей модели

```python
knn_pipe = 
([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_jobs=-1))])
```

```python
knn_params = {'knn__n_neighbors': range(1, 10)}
```

```python
knn_grid = GridSearchCV(knn_pipe, knn_params,
cv=5)
```

```python
knn_grid.fit(X_train, y_train);
```

```python
knn_grid.best_params_, knn_grid.best_score_
```

```python
accuracy_score(y_valid, knn_grid.predict(X_valid))
```

## На пятикратной кросс-валидации мы получили, что оптимальное число ближайших соседей:7. Точность модели на отложенной выборке 93%.
