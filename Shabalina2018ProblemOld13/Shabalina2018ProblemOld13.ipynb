## Площадь единичного круга
Оценить площадь единичного круга методом Монте-Карло. Построить график зависимости результата от размера выборки.

Случайным образом выбирается некоторое число точек из единичного квадрата.

$$\dfrac{\text{кол-во точек внутри круга}}{\text{объем выборки}} \approx \dfrac{S_{\text{круга}}}{S_{\text{квадрата}}} = \dfrac{\pi}{4}$$ 

Будем брать выборку объемом $100n$, где $n$ принимает значения от 1 до 499, будем строить зависимость величины 

$$4 \cdot \dfrac{\text{кол-во точек внутри круга}}{\text{объем выборки}} ~ \text{от $n$.}$$

Согласно теории результат должен колебаться около $\pi$.

![title](machine_learning_1.png)

```python
result = np.zeros(499)

for n in range (1, 500):
    
    x = np.random.uniform(0,1,100*n)
    y = np.random.uniform(0,1,100*n)
    
    
    res = np.zeros(100*n)

    for i in range (0,100*n):   
        if (np.sqrt((x[i]-0.5)**2 + (y[i]-0.5)**2) <= 0.5):
                res[i] = 1.0   
    result[n-1] = 4*(sum(res)/(100*n))
```

```python
x_period = np.linspace(1,499,499)
y_period = result
plt.plot(x_period, y_period)
plt.show()
```
